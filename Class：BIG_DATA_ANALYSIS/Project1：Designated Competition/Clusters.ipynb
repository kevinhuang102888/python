```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
import matplotlib.pyplot as plt
```

```python
data_yr1=dd.read_csv('data_yr1.csv')
chids=data_yr1.chid
data_yr1=data_yr1.iloc[:,2:69]
data_yr1.compute()
```

```python
data_yr2=dd.read_csv('data_yr2.csv')
chids2=data_yr2.chid
data_yr2=data_yr2.iloc[:,2:69]
data_yr2.compute()
```

```python
import dask_ml.cluster
model=dask_ml.cluster.KMeans(n_clusters=8)
labels=model.fit(data_yr1.to_dask_array(lengths=True))
```

```python
labels=pd.DataFrame(labels.labels_.compute())
```

```python
labels.value_counts()
```

```python
labels.columns=["cluster"]
```

```python
labels
```

```python
labels[labels["cluster"]==1]=7
```

```python
data_yr1_cluster=pd.concat([chids.compute().reset_index(),labels],axis=1).iloc[:,1:69]
```

```python
label_2=model.predict(data_yr2.to_dask_array(lengths=True).compute())
```

```python
data_yr2_cluster=pd.concat([chids2.compute().reset_index(),pd.DataFrame(label_2.compute())],axis=1).iloc[:,1:69]
```

```python
data_yr2_cluster.columns=["chid","cluster"]
```

```python
data_yr2_cluster["cluster"].value_counts()
```

```python
data_yr2_cluster["cluster"][data_yr2_cluster["cluster"]==1]=7
```

```python
data_yr1_cluster.to_csv("data_yr1_cluster.csv", index = False)
```
