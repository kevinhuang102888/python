```python
def log_data_transformation(data,column_name):
    log_data=data[column_name].apply(lambda x : np.log(x+1))
    log_data[log_data>log_data.quantile(0.999)]=log_data[log_data<=log_data.quantile(0.999)].max()
    plt.hist(log_data,bins=30)
    plt.title(column_name+", log transformation")
    plt.show()
    return log_data
```

```python
def counts_rep(data):
    x=0
    if data["top1"]==data["top2"]:
        x=x+1
    if data["top2"]==data["top3"]:
        x=x+1
    if data["top1"]==data["top3"]:
        x=x+1
    return x
```
