```python
def models(data):
    from sklearn.model_selection import train_test_split
    train_data_0,test_data_0,train_labels_0,test_labels_0=train_test_split(data.iloc[:,0:68],data['top_1_y'],test_size=0.2, random_state=2021)
    import lightgbm as lgb
    # Create the model
    LGBM_0 = lgb.LGBMClassifier(application='multiclass', boosting='gbdt', learning_rate=0.03, 
                              max_depth=15, feature_fraction=0.5, random_state=69,n_estimators=150,colsample_bytree=0.7)
    # Learn the digits on the train subset
    model=LGBM_0.fit(train_data_0, train_labels_0)
    
    # Predict the value of the digit on the testing subset
    predicted_0 = LGBM_0.predict(test_data_0)
    
    from sklearn import metrics
    print(metrics.accuracy_score(train_labels_0,LGBM_0.predict(train_data_0)))
    print(metrics.accuracy_score(test_labels_0,predicted_0))
    
    return model
```

```python
def predict_prob(model,data):
    prob =model.predict_proba(data)
    ids=np.argsort(-prob, axis=1)
    top1=[model.classes_[i[0]] for i in ids]
    top2=[model.classes_[i[1]] for i in ids]
    top3=[model.classes_[i[2]] for i in ids]
    final=pd.DataFrame({"chid":data.chid,"top1":top1,"top2":top2,"top3":top3})
    return final
```
