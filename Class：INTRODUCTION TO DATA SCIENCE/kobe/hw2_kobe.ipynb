```python
#列出Kobe 對戰過的球隊中，使得Kobe 平均得分最低的前五支球隊，並顯示平均得分。
def bottom5():
    import pandas as pd
    cobe = pd.read_csv('kobe.csv')
    newcobe = cobe[['shot_made_flag','shot_type','game_date','opponent']]
    sort=newcobe.sort_values(by=['opponent','game_date','shot_made_flag']).copy()
    opponental=sort['opponent'].unique()
    data=[]
    point2=sort['shot_type']=='2PT Field Goal'
    point3=sort['shot_type']=='3PT Field Goal'
    getpoint=sort['shot_made_flag']
    for x in opponental:
        a=sort['opponent']==x
        opponent=sort[a]['game_date'].unique()
        day=len(opponent)
        two=len(sort[a & point2 & getpoint])*2
        three=len(sort[a & point3 & getpoint])*3
        total=two+three
        data.append((total/day,x))
    data=sorted(data)
    print('NO1:',data[0][1],data[0][0])
    print('NO2:',data[1][1],data[1][0])
    print('NO3:',data[2][1],data[2][0])
    print('NO4:',data[3][1],data[3][0])
    print('NO5:',data[4][1],data[4][0])
bottom5()
```

```python
#列出Kobe 在季後賽最後3 分鐘內得分最高的前五場球賽，並顯示其得分。
def threeminute():
    import pandas as pd
    cobe = pd.read_csv('kobe.csv')
    playoffs=cobe['playoffs']==1
    period=cobe['period']==4
    minute=cobe['minutes_remaining']*60
    second=cobe['seconds_remaining'] 
    three=(minute+second)<=180
    point2=cobe['shot_type']=='2PT Field Goal'
    point3=cobe['shot_type']=='3PT Field Goal'
    getpoint=cobe['shot_made_flag']
    new=cobe[playoffs&period&three]
    date=new['game_date'].unique()
    final=[]
    for x in date:
        day=new['game_date']==x
        nowdata=new[day]
        total=len(nowdata[point2&getpoint])*2+len(nowdata[point3&getpoint])*3
        final.append((total,x))
    final=sorted(final,reverse=True)
    print('NO1:',final[0][1],final[0][0])
    print('NO2:',final[1][1],final[1][0])
    print('NO3:',final[2][1],final[2][0])
    print('NO4:',final[3][1],final[3][0])
    print('NO5:',final[4][1],final[4][0])
threeminute()
```

```python
#列出Kobe 各球季季後賽中，比賽最後1 分鐘內的Jump Shot 命中率，按球季先後排序。
def jumpshot():
    import pandas as pd
    cobe = pd.read_csv('kobe.csv')
    playoffs=cobe['playoffs']==1
    action=cobe['action_type']=='Jump Shot'
    one=(cobe['minutes_remaining']*60+cobe['seconds_remaining']) <=60
    point2=cobe['shot_type']=='2PT Field Goal'
    point3=cobe['shot_type']=='3PT Field Goal'
    getpoint=cobe['shot_made_flag']
    new=cobe[playoffs&action&one]
    season=pd.DatetimeIndex(new['game_date']).year.unique()
    final=[]
    for k in season:
        now=pd.DatetimeIndex(new['game_date']).year==k
        nowdata=new[now]
        count=len(nowdata)
        total=len(nowdata[point2&getpoint])*2+len(nowdata[point3&getpoint])*3
        final.append((k,total/count))
    final=sorted(final)
    season=pd.DataFrame(data=final,columns=['year','rate'])
    return season
jumpshot()
```

```python
#計算Kobe「得分命中率33%以上的最長連續場數」前3 名，列出場數以及起訖日期。
def thirtythree():
    import pandas as pd
    import numpy as np
    cobe = pd.read_csv('kobe.csv')
    cobe['game_date'] =pd.to_datetime(cobe.game_date)#轉換成pandas的日期，才可以排序
    cobe2 = cobe.sort_values(by=['game_date','shot_made_flag'])#由時間排序
    date=cobe2['game_date'].unique()#有多少個不重複的時間
    day=1#最初天數
    final=[]
    aldate=[]#連續的日期
    for x in date:
        aldate.append(x)
        condition=cobe2['game_date']==x#條件
        new=cobe2[condition]['shot_made_flag']#在該天的Shot_made_flag
        number=len(new)#投幾球
        total=sum(new)#進幾球
        rate=total/number
        if rate>=0.33:
            day=day+1
        else:
            final.append((day,aldate[0],aldate[-1]))#天數/第一天/最後一天
            day=1
            aldate=[]
    final=sorted(final,reverse=True)#由總天數重新排序，由最多到最少
    for x in range(0,3):#把numpy.datetime64轉換回字串
        ts1=pd.to_datetime(str(final[x][1]))
        ts2=pd.to_datetime(str(final[x][2]))
        d1=ts1.strftime('%Y-%m-%d')
        d2=ts2.strftime('%Y-%m-%d')
        print('NO%d:'%(x+1),'from',d1,'to',d2,'總共',final[x][0])
thirtythree()
```
