```python
import matplotlib.pyplot as plt 
import numpy as np
with open('cdata.csv','r') as c:
    x=[]
    y=[]
    c.readline()
    for h in c.readlines():
        h=h.strip()
        p=h.split(',')
        x.append(float(p[0]))
        y.append(float(p[1]))
```

```python
import random
plt.scatter(x,y)#一定要兩個變數
r=random.sample(range(len(x)), 4)
c=['red','blue','green','purple']
plt.scatter(x[r[0]],y[r[0]],c=c[0])
plt.scatter(x[r[1]],y[r[1]],c=c[1])
plt.scatter(x[r[2]],y[r[2]],c=c[2])
plt.scatter(x[r[3]],y[r[3]],c=c[3])
```

```python
a=[[[],[],[],[]],[[],[],[],[]]]
for n in range(0,len(x)):
    if n in r:
        continue
    else:
        distance=[(x[s]-x[n])**2+(y[s]-y[n])**2 for s in r]
        m=distance.index(min(distance))
        plt.scatter(x[n],y[n],c=c[m])
        a[0][m].append(x[n])
        a[1][m].append(y[n])
```

```python
def findcenter(a):
    newx1=sum(a[0][0])/len(a[0][0])
    newx2=sum(a[0][1])/len(a[0][1])
    newx3=sum(a[0][2])/len(a[0][2])
    newx4=sum(a[0][3])/len(a[0][3])
    newy1=sum(a[1][0])/len(a[1][0])
    newy2=sum(a[1][1])/len(a[1][1])
    newy3=sum(a[1][2])/len(a[1][2])
    newy4=sum(a[1][3])/len(a[1][3])
    plt.scatter(newx1,newy1,marker="+",color='black')
    plt.scatter(newx2,newy2,marker="+",color='black')
    plt.scatter(newx3,newy3,marker="+",color='black')
    plt.scatter(newx4,newy4,marker="+",color='black')
    return newx1,newx2,newx3,newx4,newy1,newy2,newy3,newy4
```

```python
number=0
while True:
    n1,n2,n3,n4,f1,f2,f3,f4=findcenter(a)
    for h in range(0,len(x)):
        b=[[[],[],[],[]],[[],[],[],[]]]
        first=(n1-x[h])**2+(f1-y[h])**2
        second=(n2-x[h])**2+(f2-y[h])**2
        third=(n3-x[h])**2+(f3-y[h])**2
        forth=(n4-x[h])**2+(f4-y[h])**2
        d=[first,second,third,forth]
        if m==d.index(min(d)):
            number=number+1
        m=d.index(min(d))
        plt.scatter(x[h],y[h],c=c[m])
        b[0][m].append(x[h])
        b[1][m].append(y[h])
    if number>290:
        break
plt.show()
```
